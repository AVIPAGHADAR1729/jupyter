```python
import pandas as pd
```

```python
df=pd.read_json('../data/appLog_March/01_03_2021.json')
```

```python
df['appdata'][2]
```

```python
for i in df['appdata']:
    print(i)
```

```python
app_data=pd.DataFrame(columns=['appName','pkgName','eventType','eventFlag','className','datetime','timestamp'])
```

```python

for i,j in enumerate(df['appdata']):
   
    app_data.loc[i,'appName']=j['appName']
    app_data.loc[i,'pkgName']=j['pkgName']
    app_data.loc[i,'eventType']=j['eventType']
    app_data.loc[i,'eventFlag']=j['eventFlag']
    app_data.loc[i,'className']=j['className']
    app_data.loc[i,'datetime']=j['datetime']
    app_data.loc[i,'timestamp']=j['timestamp']
    
    
    
    
    
```

```python
app_data.head()
```

```python
import seaborn as sns
```

```python
import matplotlib.pyplot as plt
plt.figure(figsize=(10,6))

plt.bar(app_data['appName'],app_data['datetime'])
plt.xticks(rotation=90)
plt.show() 
```

```python
app_data=app_data.dropna()
```
