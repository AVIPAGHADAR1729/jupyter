```python
def naive_table(freq_naive):
    pos_sum=freq_naive["label_1"].sum()
    neg_sum=freq_naive["label_0"].sum()
    v=len(freq_naive["label_0"])
    freq_naive["label_1"]=freq_naive["label_1"].apply(lambda x:(x+1)/(v+pos_sum))
    freq_naive["label_0"]=freq_naive["label_0"].apply(lambda x:(x+1)/(v+neg_sum))
    return freq_naive
    
#     for i in freq_naive['word'].values:
#         naive_table_df=pd.DataFrame(columns=["Positive","Negative"])
#         naive_table_df.iloc[i]

    
    
    
    
```

```python

compress_text_3d=pd.DataFrame(columns=["bias","positive","negative","sentiment"])
```

```python
def pos_neg(txt):
    pos=0
    neg=0
    for i in txt:
        
        if i in keys:
#             print(pos,neg)
            if (i,1) in freq.keys():
               pos=pos+freq[(i,1)]
            
            if (i,0) in freq.keys():
                
                neg=neg+freq[(i,0)]
            
    return (pos,neg)
       
        
```

```python
for i,j in enumerate(df_under_sample["process_review"]):
    compress_text_3d.loc[i,"positive"],compress_text_3d.loc[i,"negative"]=pos_neg(j)
#     print(j)
#     print(pos_neg(j))
    compress_text_3d.loc[i,"bias"]=1
    compress_text_3d.loc[i,"sentiment"]=df_under_sample.loc[i,"rating"]
    
#     break
    
```
