```python
from util1.utils import build_freqs
freq=build_freqs(list(df_under_sample['review']),df['rating'])
```

```python
freq_table=pd.DataFrame(columns=['word','label_1','label_0'])
#logic 1
# for i,j in enumerate(freq.keys()):
    
#     freq_table.loc[i,'word']=j[0]
#     try:
#         if j[1]==1:
#             freq_table.loc[i,'label_1']=freq[j]
#             freq_table.loc[i,"label_0"]=freq[(j[0],0)]

#         else:
#             freq_table.loc[i,'label_0']=freq[j]
#             freq_table.loc[i,"label_1"]=freq[(j[0],1)]
#     except:
#         freq_table.loc[i,"label_0"]=0
#         freq_table.loc[i,"label_1"]=0


    
    
```

```python
#logic 2
freq_table=pd.DataFrame(columns=['word','label_1','label_0'])
word_list=[]
for i in freq.keys():
    word_list.append(i[0])
for i,word in enumerate(word_list):
    pos=0
    neg=0
    if (word,1) in freq.keys():
        pos=freq[(word,1)]
    
    if (word,0) in freq.keys():
        neg=freq[(word,0)]
        
    if (abs(pos-neg)>100):
    
        freq_table.loc[i,'word']=word
        freq_table.loc[i,"label_1"]=pos
        freq_table.loc[i,"label_0"]=neg
    else:
        pass 
```

```python
freq_table.shape

```

```python
freq_table["log_1"]=np.log([x for x in freq_table["label_1"]])
freq_table["log_0"]=np.log([x for x in freq_table["label_0"]])
```

```python
fig=px.scatter(data_frame=freq_table,x="log_1",y="log_0",text="word",height=800)
fig.update_traces(textfont_size=20)
fig.show()
```

```python
compress_text_3d=pd.DataFrame(columns=["bias","positive","negative","sentiment"])
```

```python
freq_table.head()
```

```python
def pos_neg(txt):
    pos=0
    neg=0
    for i in txt:
        
        if i in freq_table["word"].values:
#             print(pos,neg)
            pos=pos+freq[(i,1)]
            neg=neg+freq[(i,0)]
            
    return (pos,neg)
       
        
```

```python
for i,j in enumerate(df_under_sample["process_review"]):
    compress_text_3d.loc[i,"positive"],compress_text_3d.loc[i,"negative"]=pos_neg(j)
#     print(j)
#     print(pos_neg(j))
    compress_text_3d.loc[i,"bias"]=1
    compress_text_3d.loc[i,"sentiment"]=df_under_sample.loc[i,"rating"]
    
#     break
    
```
