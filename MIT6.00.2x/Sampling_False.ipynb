```python
import statistics as st
import random as rd
import matplotlib.pylab as plt
import pandas as pd
```

```python
temp_table=pd.read_csv("../data/temperatures.csv")

temp_table.info()
k=temp_table.iloc[0:,[1]]
l1=list(k['TEMP'])
l1
```

```python

popmean = st.mean(l1)
popmean
```

```python
len(l1)
```

```python
"for random samlping"
sum=0
samplesize=200
num_bad=0
num_trials=1000
l2=[]
l3=[]
for i in range(num_trials):
    sample=rd.sample(l1,samplesize)
    sample_mean=st.mean(sample)
    l3.append(i)
    l2.append(sample_mean)
    se=st.stdev(sample)/(samplesize**0.5)
    if abs(popmean-sample_mean)>1.96*se:
        num_bad =num_bad+1
print("95% confidence outside interval",num_bad/num_trials)
         
```

```python
"for random samlping"
sum=0
samplesize=200
num_bad=0
num_trials=1000
for i in range(num_trials):
    x=range(0,len(l1)-samplesize)
    start=rd.choice(x)
    sample=l1[start:start+samplesize]
    sample_mean=st.mean(sample)
    se=st.stdev(sample)/(samplesize**0.5)
    if abs(popmean-sample_mean)>1.96*se:
        num_bad =num_bad+1
    
print("95% confidence outside interval",num_bad/num_trials)
         
```
