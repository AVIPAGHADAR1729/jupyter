```python
def datef(x):
  
    l,m,n=x[0],x[1],x[2]
    x[0]=l
    x[1]=n
    x[2]=m
    return str("20"+x[0]+"-"+x[1]+"-"+x[2])
    
def search_words(text):
    result = re.findall(r'\w+', text)
    result = result[0:3]
    return " ".join(result)

def clean_data(df):
#   trending_date
    df['trending_date'] = df['trending_date'].str.replace(".","-")
    df['trending_date'] = df['trending_date'].str.split("-").apply(lambda x:datef(x))
    df['trending_date']=pd.to_datetime(df['trending_date'])
# publish_time
    df['publish_time']=pd.to_datetime(df['publish_time'])    
    df['regular_date']=yt['trending_date'].apply(lambda x:x.date())

# making object to numeric
    df['views']=df['views'].astype('Int64')
    df['likes']=df['likes'].astype('Int64')
    df['dislikes']=df['dislikes'].astype('Int64')
    df['comment_count']=df['comment_count'].astype('Int64')
    return df
```

```python
yt = clean_data(yt)
yt['video_short_names'] = yt['title'].apply(lambda x: search_words(x))
```

# 4 Most viewed - bar plot

```python
def most_views_video(df=None):
    df=df[df['trending_date']=='2017-11-14']
    df=df.sort_values(by='views',ascending=False)[:10]
    df=df.sort_values(by='views')
    fig=px.bar(data_frame=df,width=800,
                  x='views',y='video_short_names',text='views')
    fig.update_traces(texttemplate='%{text:.3s}', textposition='outside')
    

    fig.show()
    
```

```python
most_views_video(yt)
```

```python

def most_likes_video(df=None):
    
    df=df[df['trending_date']=='2017-11-14']
    df=df.sort_values(by='likes',ascending=False)[:10]
    df=df.sort_values(by='likes')
    y_title=df.title
    fig=px.bar(data_frame=df,width=1000,
                  title="likes vs videos",
                  x='likes',y='video_short_names',text='likes')
    fig.update_traces(texttemplate='%{text:.2s}', textposition='outside')
    fig.show()
    
```

```python
def top_10_trending_videos(df):
    df['likes'] = df['likes']+1
    df['likes/views'] = df['likes']/df['views']
    df['likes/views'] = df['likes/views'].apply(lambda x: round(x, 3))
    df = df[df['likes/views'] > 0.1]
    print(df['likes/views'].unique())
    df=df.sort_values(['trending_date', 'likes/views'], ascending=[True, True])
    df['t_date'] = df['trending_date'].astype(str)
    df.sort_values(by='trending_date')
    fig = px.bar(df, x='likes/views', y='video_short_names',
                 hover_data={'video_short_names': True},
                 title='Time Series with Rangeslider',
                 animation_frame='t_date',
        
                 color='likes/views',
                 color_continuous_scale=px.colors.sequential.haline,

                 text='likes/views',
                 orientation='h', range_x=(0, 0.6),
                 )
    fig.update_layout(margin=dict(t=40, b=0, l=0, r=0))
    
    fig.update_yaxes(automargin=True)


    fig.layout.updatemenus[0].buttons[0].args[1]["frame"]["duration"] = 5000
    fig.layout.updatemenus[0].buttons[0].args[1]["transition"]["duration"] = 5000
#     fig.update_layout(updatemenus=[  dict(
#             type="frame1",
#             direction="right",

#             showactive=True)])
    fig['layout']['updatemenus'][0]['pad'] = dict(r=5, t=80)
    fig['layout']['sliders'][0]['pad'] = dict(r=5, t=60)
    fig.update_traces(textposition='auto')
#     fig['layout']['sliders'][0]['transition']= {'duration': 300, 'easing': 'exp'}


#     fig.update_xaxes(rangeslider_visible=True)
    fig.show()


top_10_trending_videos(yt)
```
